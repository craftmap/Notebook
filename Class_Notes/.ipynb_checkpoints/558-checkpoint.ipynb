# Chapter 8 Reading

Dislocations
- Edge 
    - Cause tension below them and so solute atoms tend to gather there to help relieve stress
- Screw
    - Two of these will have an attractive force on each other

- Energy_Edge ~ 1.5x E_screw
- Energy is proportional to dislocation length and b^2 (Burgers vector)
- Franks rule: Dislocations reaction is favorable if b1^2+b2^2>b3^3
    - This favors smaller B vectors
- Dislocations can only move in plane that contains both dislocation line and burgers vector
    - For a screw these are parallel and so they can "Cross Slip" and change planes
- Staking fault energy (SFE)
    - Fig 8.23: Low SFE metals are less likely to experience Cross Slip and therefor have straight slip lines compared to wave ones in high SFE metals
    
    
# Chapter 9 Reading

Annealing

- Stages of Annealing (In order of increasing Time & Temperature)
    1) Recovery- Rearrangement of dislocations to form subgrains
    2) Recrystallization- Replacement of cold-worked grains by new ones (major hardness decrease)
    3) Grain Growth
    
- Recovery
    - Driving Forces:
        - decrease in vacancy concentration
        - energy reduction from moving dislocations to boundaries
        - relieving stored elastic energy
    - Energy released is usually from annealing out of point defeccts and rearranging dislocations
    - Electrical resistivitiy drops
    - SUMMARY: In recovery there is a large decrease in the number of point defects as they diffuse to edge dislocations and grain boundaries. Dislocations rearrange themselves into lower energy configurations leading to the formation of subgrains. Enough creep occurs to reduce elastic strains. There are major changes in electrical properties, some softening, and a reduction of residual stresses, but no changes of microstructure."
    
- Recrystallization
    - Cold-worked grains with high dislocation deinsity are replaced by new grains with much lower dis. en.
    - T_recrystallization: Temp at which 50% of sample recrystalizes in 30 minutes
    - More cold work and also smaller grains lead to lower T_recry
    - More cold work increases both nucleation rate and growth rate which decreases grain sizes formed by recry.
    - Impurities slow grwth rate which leads to more nucleation and recry graims are therefore smaller
    - ?? True or False (as books says?)
        1) Recry grains are rarely equiaxed
        2) Orientation of recry. grains are not random
- Grain Growth
    - Reduction of energy through reduction of grain boundaries
        1) Boundaries tend to move toward center of curvature
        2) Dihedral angles between three grain boundaries average 120deg
    - Large grains are concave in and grow out
        - Small grains are convex and are eaten by big ones
    - Rate of growth is proportional to 1/r_curv
    
- Grain Size limit
    - Impurities retard growth rate
    - Size is limited by the second-phase particles that impede growth by the drag they cause
        - D_max = (4/3)r/f where f is volume fraction of second phase EQ 9.8
    - Bamboo Structure: For wire D_grain_max = 1.5* D_wire 
    - Coarse grains impart creep resistance at high temp
    - At low temp fine grains result in higher hardness and toughness
