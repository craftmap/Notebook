Where 0.01 is a typical cell size in that area

# FINAL SETUP FOR ADAPTIVE MESHING
## Field Functions
$$\$\$GradV = grad(mag(\$ \$ Velocity))$$

$$ \$CellWidth = pow($Volume,1/3) $$

$$\$LengthGradV = $CellWidth* mag(\$\$GradV)$$

$$\$Refine1 = ($LengthGradV>0.6)?$CellWidth/2.0:$CellWidth$$

$$\$Fill\_gap = ($LengthGradV>-9999999999)?$CellWidth:0.01$$ (.002 for rotating)


$$\$Refine2 = ($WallDistance<0.00085)?0:($LengthGradV>0.6)?$CellWidth/2.0:0$$


## Procedure
1) Increase the inner iteration steps to 9999 (Except on initial refinement)

2) Run the simulation till field is steady

3) Extract XYZ Table 2

4) Execute Rotating mesh

5) Decrease inner iterations back down (~3)

6) Run simulation til steady

7) Extract XYZ Table 1

8) Execute Fluid mesh

9) Step 1 forward

10) Change XYZ table 1 to Fill_gap

11) Extract table and remesh Fluid

12) Reset table scalar to Refine1

13) Repeat

Steps 9-12 only need to be done everytime for the convergence study. Otherwise just do them at the end. Also it may be more efficient to just run all refinement steps on the interior rotating region first and then move to rotating the outside Fluid region.
