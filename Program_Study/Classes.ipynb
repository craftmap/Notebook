## Completed
- ME EN 512: Intermediate Fluid Dynamics
    - Fluid transport properties, review of integral analysis, Navier-Stokes equations, exact and similarity solutions, boundary layers, vorticity, jets and wakes
    
- ME EN 595R (515): Aerodynamics
    
    
- ME EN 534 - Dynamics of Mechanical Systems (As Undergrad)
    - Hamiltonian and Lagrangian dynamics, generalized coordinates, linear and angular momentum, Euler angles, rigid-body motions, and gyroscopic effects. Theory taught with applications integrated.

## Potential Classes
- ME EN 561: Fundamentals of Acoustics
     - Offered: Fall
     - Description:
        - Vibrating systems, elastic media, mechanical energy, and radiation. Sound generation, transmission, reflection, and reception.
    - Prerequisits (Skipped)
        - Physc 318: Partial differential equations, classical field equations, algebra of complex variables, Fourier analysis, integral transforms, and orthogonal functions.


- ME EN 507(Me En-CE En): Linear Finite Element Methods
    - Offered:
    - Description:
        - Linear static finite element analysis of elastic solids; problem formulation, finite element discretization, and equation solving; stability, interpolation theory, and error estimates. Introduction to underlying continuum mechanics principles. Isogeometric analysis as a generalization of classical finite element analysis; basic principles of analysis-suitable computational geometry.

- ME EN 558: Metallurgy
    - Offered: Every Other Fall
    - Description: Fundamental principles of physical metallurgy and their application to design.
        - Course Outcomes: 
            - Describe common material characterization techniques and their strengths and limitations.; Describe the interactions of point and line defects and their effects on mechanical properties.; Be able to draw, describe and use phase diagrams associated with commonly used engineering materials.; Describe/model the evolution of microstructure under non-steady state conditions.; Model non-steady state diffusion in solid materials and predict the difference as a function of time, temperature and species.; Describe the thermodynamics and prevalence of both homogeneous and heterogeneous nucleation.; Describe the effect of nucleation and growth kinetics on isothermal transformations.; Describe the theory behind the different hardening mechanisms of solid materials.
